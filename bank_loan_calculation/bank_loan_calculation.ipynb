```python
features = loans_2007.drop("loan_status", axis=1)
```

```python
target = loans_2007["loan_status"]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_predict
```

```python
lr = LogisticRegression(class_weight="balanced")
```

```python
predictions_2 = cross_val_predict(lr, features, target, cv=3)
```

```python
predictions_2 = pd.Series(predictions_2)
```

```python
predictions_2.value_counts()
```

```python
fp_filter = (predictions_2 == 1) & (target == 0)
fp = len(predictions_2[fp_filter])

tp_filter = (predictions_2 == 1 ) & (target == 1)
tp = len(predictions_2[tp_filter])

fn_filter = (predictions_2 == 0) & (target == 1)
fn = len(predictions_2[tp_filter])

tn_filter = (predictions_2 == 0) & (target ==0)
tn = len((predictions_2[tn_filter]))
```

```python
tpr = tp / (tp + fn)
fpr = fp / (fp + tn)

print(tpr)
print(fpr)
```

```python
penalty = {
    0:10,
    1:1
}
lr = LogisticRegression(class_weight = penalty)

predctions = cross_val_predict(lr, features , target , cv=3)
predictions = pd.Series(predictions)

fp_filter = (predictions == 1) & (target ==0)
fp = len(predictions[fp_filter])

tp_filter = (predictions == 1) & (target ==1)
tp = len(predictions[tp_filter])

fn_filter = (predictions == 0) & (target ==1)
fn = len(predictions[fn_filter])

tn_filter = (predictions == 0) & (target ==0)
tn = len(predictions[tn_filter])

tpr = tp / (tp + fn)
fpr = fp / (fp + tn)

print(tpr)
print(fpr)
```
