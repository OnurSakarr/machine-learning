```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('/content/Churn_Modelling.csv')
```

```python
df.head(4)
```

```python
df=df.drop(columns=["RowNumber","CustomerId","Surname"])
```

```python
df.head(3)
```

```python
df=pd.get_dummies(df,columns=["Geography","Gender"],drop_first=True)
```

```python
df.head(3)
```

```python
y=df['Exited']
x=df.drop("Exited",axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=61,train_size=0.7)
```

```python
rf=RandomForestClassifier()
model=rf.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
x.head(2)
```

```python
model.predict([[607,27,2,0,1,1,1,100000,True,False,False]])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.read_csv('/content/Churn_Modelling.csv')
df=df.drop(columns=["RowNumber","CustomerId","Surname"])
```

```python
df.head(3)
```

```python
ohe=OneHotEncoder()
```

```python
ohe_data=ohe.fit_transform(df[['Geography',"Gender"]]).toarray()
```

```python
ohe.get_feature_names_out()
```

```python
ohe_data=pd.DataFrame(ohe_data)
```

```python
ohe_data.columns=ohe.get_feature_names_out()
```

```python
ohe_data
```

```python
df.head(5)
```

```python
df=df.drop(columns=["Geography","Gender"])
```

```python
df.head(5)
```

```python
df[ohe_data.columns]=ohe_data
```

```python
df.head(5)
```

```python
y=df['Exited']
x=df.drop("Exited",axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=61,train_size=0.7)
```

```python
rf=RandomForestClassifier()
model=rf.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
